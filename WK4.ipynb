```python
# Solution to HL tasks
import pandas as pd
df = pd.read_csv('Hotel data.csv')
df['Score'] = df['Average_star_rating'] * 2 # The score was obtained by multiplying 2 by the average_star_rating
df
```

```python
# Practical task 1 solution
# To determine the number of rows & columns, we use 'variable name.shape' command

df.shape    # Row = 15 & column = 5
```

```python
# Practical task 2(i) solution
df.loc[3:8]     # variable name.loc[3:8] prints all elements from row 3 through 8
```

```python
# Practical task 2(ii) solution
df.iloc[3:8]    # variable name.iloc[] prints all elements from row 3 through 7, exempting elements in row 8
```

```python
# Practical task 3 solution

df['All-Inclusive Hotel'].mean()    # variable name['column name'].mean() prints the specified mean of that variable
```

```python
# Practical task 4 solution
# In finding the lowest scoring destinations, we create a conditional statement & assign it to a variable

filter_score = df['Score'] < 9.0
filter_score.head(15)

lowscore = df[filter_score]
lowscore
```

```python
# Practical task 5 solution
# In finding the highest scoring destinations, we create a conditional statement & assign it to a variable

filter_score1 = df['Score'] > 9.0
filter_score1.head(15)

highscore = df[filter_score1]
highscore
```

```python
# Practical task 6 solution

all_inclusive_hotels = df['All-Inclusive Hotel'] > 9    # Sets a condition that includes only destinations that had >9 all-inclusive                                                            hotels
all_inclusive_hotels.head(15)

dest = df[all_inclusive_hotels]
dest
```

```python
# Practical task 7 solution

data = df['Score'] > 8  # The conditional statement filters any data above 8 and prints out the result
data.head(15)

above = df[data] 
above
```

```python
# Practical task 8 solution

data_score = df['Score'] < 2
data_score

below = df[data_score]
below
```
