```python
# Source code for task 1

def menuOrder(quantity, item_list):            # The parameters quantity and item_list must be defined with the function menuOrder
    global total                               # global keyword is used to include the variable 'total' inside a local variable
    if quantity > item_list[2]:                # item_list[n]: n= range of items in each list (0-2)
        print("There is not enough stock!")
        pass
    else:
        total += item_list[1] * quantity
        item_list[2] -= quantity

total = 0                                      # total must be initialised to zero
A = ['Burger', float(1.50), 40], ['Fries', float(0.50), 150], ['Coca cola', float(2.00), 30], ['Vegetable salad', float(1.20), 70]
        
print("Welcome to Joe's Corner")               # Number of items in stock is given as 40, 150, 30 and 70 respectively

while True:
    choice =  input("What would you like?").upper()       
    quantity =  int(input('How many would you like?'))

    if choice == 'BURGER':
      menuOrder(quantity, A[0])
      print('[1]', A[0][0:2])
    
    elif choice == 'FRIES':
        menuOrder(quantity, A[1])
        print('[2]', A[1][0:2])

    elif choice == 'COCA COLA':
        menuOrder(quantity, A[2])
        print('[3]', A[2][0:2])

    elif choice == 'VEGETABLE SALAD':
        menuOrder(quantity, A[3])
        print('[4]', A[3][0:2])
    
    else:
        print("Item not found")

    more_items = (input("Do you want to order more items?")).lower()    #.lower() is also included to check for case sensitive strings
    if more_items == "yes":
        pass
    else:
        break

print('Thank you for ordering!\nYour total cost is: Â£' +  str(total))
```

```python
# Solution to task 2

def depValue(new_year,new_cost):
    depreciation = 0.9

    while new_cost >= 1000:
        new_cost = round(new_cost * depreciation)
        new_year += 1
        print('In the year',new_year,'the cost of the motorbike will be',new_cost, 'and will be depreciating at the rate of 10% pa')

cost = int(input('Enter the cost: '))
year = int(input('Enter the year: '))


depValue(new_year=year,new_cost=cost)
```

```python
# Solution to task 3
def add(x,y):
     return x + y

def subtract(x,y):
    return x - y

def multiply(x,y):
    return x * y

def divide(x,y):
    return x / y

def exponent(x,y):
    return x ** y

def modulo(x,y):
    return x % y

print('Select operation')
print('1. Add')
print('2. Subtract')
print('3. Multiply')
print('4. Divide')
print('5. Raised to the power of')
print('6. Modulo')

while True:
    choice = input('Enter your operation(1/2/3/4/5/6): ')

    if choice in ('1','2','3','4','5','6'):
        first_num = float(input('Enter your first number: '))
        sec_num = float(input('Enter your second number: '))

        if choice == '1':
            print(first_num,'+',sec_num,'=',add(first_num,sec_num))

        elif choice == '2':
            print(first_num,'-',sec_num,'=',subtract(first_num,sec_num))

        elif choice == '3':
            print(first_num,'*',sec_num,'=',multiply(first_num,sec_num))

        elif choice == '4':
            print(first_num,'/',sec_num,'=',divide(first_num,sec_num))

        elif choice == '5':
            print(first_num,'**',sec_num,'=',exponent(first_num,sec_num))

        elif choice == '6':
            print(first_num,'%',sec_num,'=',modulo(first_num,sec_num))

        else:
            print('You have entered an invalid operation. Please try again')

```
